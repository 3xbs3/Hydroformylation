```python
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, WhiteKernel, Matern, RationalQuadratic
from sklearn import svm
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import  cross_val_score
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import ExtraTreesRegressor
from rdkit import Chem
from rdkit.Chem import AllChem
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import xgboost as xgb
import lightgbm as lgb
import matplotlib
import warnings
import time
import os

warnings.filterwarnings("ignore")

    
def R2(y_pred, y_true):       
    
    return 1- mean_squared_error(y_true,y_pred)/ np.var(y_true)
    
def scale(data, method):

    '''
    Arg:
        data:data to be processed
        method: scaler's method(include standard, minmax)
    Return:
        processed data
    '''

    sca_data = []

    if method == 'standard':
        scaler = StandardScaler()
        sca_data.append(scaler.fit_transform(data))

    if method == 'minmax':
        scaler = MinMaxScaler()
        sca_data.append(scaler.fit_transform(data))

    return sca_data

def get_fp(X):
    
    index = 167
    fp_record = []
    for i in range(len(X.iloc[:,1])):
        for j in [0, 2]:
            b = AllChem.GetMACCSKeysFingerprint(Chem.MolFromSmiles(X.iloc[i,j])).ToList()
            fp_record.append(b)
    input1 = np.array(fp_record).reshape(-1,167*2)
    input_pca1 = StandardScaler().fit_transform(input1[:, :index])
    input_pca2 = StandardScaler().fit_transform(input1[:, index:2*index])
    
    return input_pca1, input_pca2

def input_generate(df, des, global_type=None, local_type=None, reduction=None, reduct_dim=4):
    
    if global_type == None:
        olefin_global = pd.read_csv(f'./{global_type}-olefin.csv', index_col=0)
        ligand_global = pd.read_csv(f'./{global_type}-Ligand-M.csv', index_col=0)
        ligand_local = None
        olefin_local = None
    elif local_type == None:
        olefin_local = pd.read_csv(f'./{local_type}-olefin.csv', index_col=0)
        ligand_local = pd.read_csv(f'./{local_type}-Ligand-M.csv', index_col=0)
        ligand_global = None
        olefin_global = None
    elif global_type == 'maccs' and local_type != None:
        olefin_global, ligand_global = get_fp(df)
        olefin_local = pd.read_csv(f'./{local_type}-olefin.csv', index_col=0)
        ligand_local = pd.read_csv(f'./{local_type}-Ligand-M.csv', index_col=0)
    elif global_type == 'maccs' and local_type == None:
        olefin_global, ligand_global = get_fp(df)
        olefin_local = None
        ligand_local = None
    else:
        olefin_global = pd.read_csv(f'./{global_type}-olefin.csv', index_col=0)
        ligand_global = pd.read_csv(f'./{global_type}-Ligand-M.csv', index_col=0)
        olefin_local = pd.read_csv(f'./{local_type}-olefin.csv', index_col=0)
        ligand_local = pd.read_csv(f'./{local_type}-Ligand-M.csv', index_col=0)
              
    y = np.array(des)[:,-1]
    
    if reduction != None:
        if reduction == 'pca':
            pca = PCA(n_components=reduct_dim)
            
        elif reduction == 'tsne':
            pca = TSNE(n_components=reduct_dim, random_state=0)
            
        input_pca1 = pca.fit_transform(StandardScaler().fit_transform(olefin_global))
        input_pca2 = pca.fit_transform(StandardScaler().fit_transform(olefin_local))
        input_pca3 = pca.fit_transform(StandardScaler().fit_transform(ligand_global))
        input_pca4 = pca.fit_transform(StandardScaler().fit_transform(ligand_local))
        input2 = StandardScaler().fit_transform(des.iloc[:, :-1])
        input = np.hstack([input_pca1, input_pca2, input_pca3, input_pca4, input2])
    else:
        input_pca1 = StandardScaler().fit_transform(olefin_global)
        input_pca2 = StandardScaler().fit_transform(olefin_local)
        input_pca3 = StandardScaler().fit_transform(ligand_global)
        input_pca4 = StandardScaler().fit_transform(ligand_local)
        input2 = StandardScaler().fit_transform(des.iloc[:, :-1])
        input = np.hstack([olefin_global, olefin_local, ligand_global, ligand_local, input2])
 
    return input, y

def train_and_test(X, y, method, seed, df, train_size):

    # Random index
    np.random.seed(seed)
    permuted_indices = np.random.permutation(np.arange(df.shape[0]))
    # Train set: 80 % of data
    train_index = permuted_indices[: int(df.shape[0] * train_size)]
    # Test set: 20 % of data
    test_index = permuted_indices[int(df.shape[0] * train_size) :]
    
    X_train, X_test, y_train, y_test = X[train_index], X[test_index], y[train_index], y[test_index]

    test_mae_record, test_r2_record, test_mse_record = [], [], []
    valid_mse = []
    name = []  
    model_summary = []
    
    if method == 'rf':
        count = 0
        for i in [100, 150, 200]:
            for j in [9, 10, 11, 12]:
                for k in [0, 1, 2, 3]:
                    count += 1
                    name.append(f"hyper1-{i}-hyper2-{j}-hyper3-{k}")
                    model = RandomForestRegressor(n_estimators=i, max_depth=j, random_state=k)
                    model_summary.append(model)
                    valid_mse.append(np.abs(np.average(cross_val_score(estimator=model
                                , X=X_train, y=y_train, scoring='neg_mean_squared_error', cv=cv))))
    if method == 'et':
        count = 0
        for i in [100, 150, 200]:
            for j in [8, 9, 10, 11]:
                for k in [0, 1, 2, 3]:
                    count += 1
                    name.append(f"hyper1-{i}-hyper2-{j}-hyper3-{k}")
                    model = ExtraTreesRegressor(n_estimators=i, max_depth=j, random_state=k)
                    model_summary.append(model)
                    valid_mse.append(np.abs(np.average(cross_val_score(estimator=model
                                , X=X_train, y=y_train, scoring='neg_mean_squared_error', cv=cv))))
    if method == 'svr':
        count = 0
        for i in ['poly', 'rbf', 'sigmoid']:
            for j in [1, 10, 100]:
                for k in [0.001, 0.01, 0.01]:
                    count += 1
                    name.append(f"hyper1-{i}-hyper2-{j}-hyper3{k}")
                    model = svm.SVR(kernel=i, degree=3, gamma='scale', coef0=0.0, tol=0.001, C=j
                            , epsilon=k, shrinking=True, cache_size=200, verbose=False, max_iter=-1)
                    model_summary.append(model)
                    valid_mse.append(np.abs(np.average(cross_val_score(estimator=model
                                , X=X_train, y=y_train, scoring='neg_mean_squared_error', cv=cv))))
    if method == 'nn':
        count = 0
        for i in [512, 256]:
            for j in [256, 128, 64]:
                count += 1
                name.append(f"hyper1-{i}-hyper2-{j}")
                model = MLPRegressor(hidden_layer_sizes=(i,j), max_iter=500
                                ,solver='adam',activation='relu')
                model_summary.append(model)
                valid_mse.append(np.abs(np.average(cross_val_score(estimator=model
                            , X=X_train, y=y_train, scoring='neg_mean_squared_error', cv=cv))))
    if method == 'xgboost':
        count = 0
        for i in [0.01, 0.05, 0.1]:
            for j in [5, 7, 9]:
                for k in [100, 200, 300]:
                    count += 1
                    name.append(f"hyper1-{i}-hyper2-{j}-hyper3-{k}")
                    model = xgb.XGBRegressor(base_score=0.5, booster='gbtree', colsample_bylevel=1
                                    , colsample_bynode=1, colsample_bytree=1, gamma=0, gpu_id=-1
                                    , importance_type='gain',interaction_constraints=''
                                    , learning_rate=i, max_delta_step=0, max_depth=j
                                    , min_child_weight=1, missing=np.nan
                                    , n_estimators=k, num_parallel_tree=1, random_state=0
                                    , reg_alpha=0, reg_lambda=1, scale_pos_weight=1
                                    , subsample=1, tree_method='gpu_hist', validate_parameters=1, verbosity=None
                                    )
                    model_summary.append(model)
                    valid_mse.append(np.abs(np.average(cross_val_score(estimator=model
                            , X=X_train, y=y_train, scoring='neg_mean_squared_error', cv=cv))))
    if method == 'dtr':
        count = 0
        for i in [6, 7, 8, 9, 10, 11, 12, 13]:
            for j in [2, 3, 4]:
                for k in [1, 2, 3, 4]:
                    count += 1
                    name.append(f"hyper1-{i}-hyper2-{j}-hyper3-{k}")
                    model = DecisionTreeRegressor(criterion='squared_error', splitter='best', max_depth=i, min_samples_split=j
                                            , min_samples_leaf=k, min_weight_fraction_leaf=0.0, max_features=None
                                            , max_leaf_nodes=None, min_impurity_decrease=0.0
                                            , ccp_alpha=0.0)
                    model_summary.append(model)
                    valid_mse.append(np.abs(np.average(cross_val_score(estimator=model
                                , X=X_train, y=y_train, scoring='neg_mean_squared_error', cv=cv))))
    if method == 'lgbm':
        count = 0
        for i in [0.01, 0.05, 0.1]:
            for j in [3, 4, 5, 6]:
                for k in [100, 200, 300, 400]:
                    count += 1
                    name.append(f"hyper1-{i}-hyper2-{j}-hyper3-{k}")
                    model = lgb.LGBMRegressor(learning_rate=i, max_depth=j, n_estimators=k, verbosity=-1
                                    )
                    model_summary.append(model)
                    valid_mse.append(np.abs(np.average(cross_val_score(estimator=model
                            , X=X_train, y=y_train, scoring='neg_mean_squared_error', cv=cv))))
    if method == 'gpr':
        count = 0
        for i in [RBF() + WhiteKernel(), Matern(nu=2.5)+WhiteKernel(), Matern(nu=1.5)+WhiteKernel()
                    , RationalQuadratic(alpha=1) + WhiteKernel(), RationalQuadratic(alpha=2) + WhiteKernel()
                    , RationalQuadratic(alpha=0.5) + WhiteKernel()]:#, DotProduct()+WhiteKernel()
            count += 1
            name.append(f"hyper1-{i}")
            model = GaussianProcessRegressor(kernel=i)
            model_summary.append(model)
            valid_mse.append(np.abs(np.average(cross_val_score(estimator=model
                    , X=X_train, y=y_train, scoring='neg_mean_squared_error', cv=cv))))
     
    # print(np.argmin(valid_mse))
    # print(valid_mse[np.argmin(valid_mse)])
    # print(name[np.argmin(valid_mse)])

    # Performance
    clf = model_summary[np.argmin(valid_mse)]
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    test_mae_record.append(mean_absolute_error(y_test, y_pred))
    test_r2_record.append(R2(y_test, y_pred))
    test_mse_record.append(mean_squared_error(y_test, y_pred))
    print(f'mae: {test_mae_record}, r2: {test_r2_record}, mse: {test_mse_record}')
    
    return y_test, y_pred

if __name__ == '__main__':
    
    # Hyperparameters
    root_dir = './'
    test_size = 0.2
    seed = 1
    cv = 5
    
    # Record the time
    start = time.perf_counter()

    # Load data
    des = pd.read_csv(os.path.join(root_dir, 'descriptor.csv'), index_col=0)
    df = pd.read_csv(os.path.join(root_dir, 'hydroformylation.csv'), index_col=0)
    X, y = input_generate(df, des, global_type='maccs', local_type='acsf', reduction='tsne', reduct_dim=3)
    
    # Train and test
    y_test, y_pred = train_and_test(X, y, 'et', seed, df, train_size=0.8)
    
    end = time.perf_counter()
    print(f"model get feature time consumed: {np.round(end - start, decimals=3)} sec.")
```
